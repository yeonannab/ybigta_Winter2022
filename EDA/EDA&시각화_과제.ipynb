### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
df_num = df.isnull().sum().to_frame()
df_num.rename(columns={0:'missing_num'}, inplace=True)

df_ratio = df.isnull().sum().to_frame() / len(df) * 100
df_ratio.rename(columns={0:'missing_%'}, inplace=True)

df_null = pd.concat([df_num, df_ratio], axis=1)
df_null = df_null[(df_null['missing_num'] > 0)]
df_null = df_null.sort_values(by='missing_num', ascending=False)
display(df_null)

# missingRatio = round(df.isnull().sum() / len(df) * 100, 1)
# missingRatio = missingRatio[missingRatio>0]
# df_null = pd.DataFrame(missingRatio, columns=['missing_ratio'])
# df_null = df_null.sort_values(by='missing_ratio', ascending = False)
# display(df_null)
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
plt.figure(figsize = (15,4))
df_null = df_null.reset_index()
# df_null
sns.barplot(x = 'index', y = 'missing_%', data = df_null)
```

- #### <font color='orange'> Q4. 결측치가 적은 beds 컬럼을 좀 더 자세히 살펴보겠습니다.

- #### <font color='orange'> Q6. 위 그래프의 세로축이 갯수가 아닌 비율을 나타내도록 재구성해보세요.

```python
# df_room_type_num = df['room_type'].value_counts().to_frame()
# df_room_type_num
plt.figure(figsize = (10,7))
df_room_type_ratio = df['room_type'].value_counts().to_frame() / len(df['room_type']) * 100
df_room_type_ratio.rename(columns={'room_type':'room_type_%'}, inplace=True)
#  df_room_type_ratio
df_room_type_ratio = df_room_type_ratio.reset_index()
sns.barplot(x = 'index', y = 'room_type_%', data = df_room_type_ratio)
```
