#### quailty가 7 이상인 데이터 추출 (20개)

```python
wine_quality_7plus = wine[wine['quality'] >= 7]
wine_quality_7plus.head(20)
```

#### alcohol을 기준으로 10도가 넘으면 'strong' 아니면 'weak'으로 표시 (lambda 쓰세요)

```python
wine['power'] = wine['alcohol'].apply(lambda x: 'strong' if x > 10 else 'weak')
wine.head()
```

#### merge 실습을 위해 데이터를 인위적으로 나눠볼게요

```python
df_x = wine[['quality', 'fixed acidity', 'citric acid']]
```

```python
df_y = wine[['quality', 'volatile acidity', 'chlorides', 'alcohol']]
```

```python
df_y = df_y[df_y.quality >= 7]
```

#### quality를 기준으로 inner join

```python
pd.concat([df_x, df_y], join='inner').head()
```

#### quality를 기준으로 outer join

```python
pd.concat([df_x, df_y]).head()
```
